```python
DF=p.read_table("{}.csv".format(sl[500]),sep=',', index_col=0)#change index of sl to access files of different shares, total index=501 (0-500) 
DF['percentage']=DF['close-adj'].pct_change()*100
```

```python
# X = DF.iloc[:,:5].values  
# y = DF.iloc[:,7].values   
import datetime as dt
DF['date'] = p.to_datetime(DF['date'])
DF['date']=DF['date'].map(dt.datetime.toordinal)
DF
DF.percentage.iloc[0]=0
DF
```

```python
X = DF.iloc[:,7].values  
y = DF.iloc[:,3].values 
```

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)  
X_train=X_train.reshape(-1, 1)
X_test=X_test.reshape(-1, 1)
X_train
```

```python
from sklearn.linear_model import LinearRegression  
regressor = LinearRegression()  
regressor.fit(X_train, y_train) 
```

```python
regressor.intercept_
```

```python
regressor.coef_
```

```python
y_pred = regressor.predict(X_test)
```

```python
df = p.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df
```

```python
from sklearn import metrics  
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', nmp.sqrt(metrics.mean_squared_error(y_test, y_pred))) 
```
